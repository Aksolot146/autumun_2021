```python
from tensorflow.keras.applications.resnet_v2 import ResNet101V2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet_v2 import preprocess_input, decode_predictions
import numpy as np
```

**Загружаем сеть**

```python
model = ResNet101V2(weights = 'imagenet')
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

**Загружаем фото**

```python
img = image.load_img('/content/drive/My Drive/horse1.jpg', target_size=(224, 224, 3))
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
img
```

**Выполняем распознавание объекта**

```python
preds = model.predict(x)
```

**Печатаем 3 наиболее вероятных предсказания**

```python
print('Результаты распознавания: ', decode_predictions(preds, top=3)[0])
```
