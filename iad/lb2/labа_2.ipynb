```python
import shutil
import os
```

```python
#Создаем Каталог с набором данных
data_dir = '/content/PetImages/'
#Каталог с данными для обучения
train_dir = 'train'
#Каталог с данными для проверки
val_dir   = 'val'
#Каталог с данными для тестирования
test_dir  = 'test'
#Часть набора данных для тестривания
test_portion = 0.15
#Часть набора данных для проверки
val_portion = 0.15
#Количество элементов данных в одном классе
nb_images = 12500
```

```python
def create_dir(dir_name):#создаем директории
  
  if os.path.exists(dir_name):
    shutil.rmtree(dir_name)

  os.makedirs(dir_name)
  os.makedirs(os.path.join(dir_name, "cats"))
  os.makedirs(os.path.join(dir_name, "dogs"))
```

```python
create_dir(train_dir)
create_dir(test_dir)
create_dir(val_dir)
```

```python
def copy_images(st, ed, src, dst):   #копируем фотографии из архива
  for i in range(st, ed):
    shutil.copy2(os.path.join(src, "Cat/" + str(i) + ".jpg"), os.path.join(dst, "cats"))
    shutil.copy2(os.path.join(src, "Dog/" + str(i) + ".jpg"), os.path.join(dst, "dogs"))
```

```python
start_val = int(nb_images * (1 - val_portion - test_portion)) #делим их на обучающие, валидационные и тестовые выборки
start_tst = int(nb_images * (1 - test_portion))
print(start_val)
print(start_tst)
# train - 0..8749
# val   - 8750 .. 10624
# test  - 10625 .. 12500
```

```python
# train
copy_images(0, start_val, data_dir, train_dir)#копируем изображения
```

```python
copy_images(start_val, start_tst, data_dir, val_dir)
```

```python
copy_images(start_tst, nb_images, data_dir, test_dir)
```

```python
img_width, img_height = 150, 150 
input_shape = (img_width, img_height, 3)
epochs = 2
batch_size = 128 #размер минивыборки
nb_train = 17500 
nb_val   = 3750
nb_test  = 3750
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
datagen = ImageDataGenerator(rescale = 1./255)
```

```python
train_generator = datagen.flow_from_directory(train_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')#загружаем в train
```

```python
val_generator = datagen.flow_from_directory(val_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')#через генератор
```

```python
test_generator = datagen.flow_from_directory(test_dir, 
                                              target_size=(img_width, img_height),
                                              batch_size = batch_size,
                                              class_mode='binary')
```

```python
%tensorflow_version 2.x
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input, decode_predictions
import numpy as np
```

```python
vgg19_net = VGG19(weights = 'imagenet', include_top = False, input_shape=(150, 150, 3))
vgg19_net.trainable=False 
```

```python
from tensorflow.keras.models import Sequential
from keras.layers.convolutional import Conv2D, MaxPooling2D
from tensorflow.keras.layers import Dropout, Flatten, Activation, Dense
from keras.preprocessing.image import ImageDataGenerator
```

```python
model = Sequential()
model.add(vgg19_net)
model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.summary()#выводим модель
```

```python
from tensorflow import keras
opt = keras.optimizers.Adam(learning_rate = 1e-5)#ограничение чтобы не возникло переобучение
model.compile(loss = 'binary_crossentropy', optimizer = opt, metrics=['accuracy'])
```

```python
model.fit_generator(train_generator, 
                    steps_per_epoch = nb_train // batch_size,
                    epochs = epochs,
                    validation_data  = val_generator,
                    validation_steps = nb_val // batch_size)
```

```python
scores = model.evaluate_generator(test_generator, nb_test // batch_size) #точность
```

```python
print(scores)
```
