```python
from tensorflow.keras.applications.densenet import DenseNet169
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.densenet import preprocess_input, decode_predictions
import numpy as np
```

```python
# Загружаем сеть
model = DenseNet169(weights = 'imagenet')
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
img = image.load_img('/content/drive/My Drive/white_horse.jpg', target_size=(224, 224, 3))
```

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
#Выполняем распознавание объекта
preds = model.predict(x)
```

```python
# Печатаем 3 наиболее вероятных предсказания
print('Результаты распознавания: ', decode_predictions(preds, top=3)[0])
```
