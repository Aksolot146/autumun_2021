Сеть ResNet50 в Keras

```python
%tensorflow_version 2.x
from tensorflow.keras.applications.resnet50 import ResNet50	
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50	 import preprocess_input, decode_predictions
import numpy as np
```

Загружаем сеть

```python
model = ResNet50(weights = 'imagenet')
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

Загружаем изображение

Лошадь

```python
img = image.load_img('/content/drive/My Drive/конь.jpg', target_size=(224, 224))
```

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

Выполняем распознавание объекта

```python
preds = model.predict(x)
```

Печатаем 3 наиболее вероятных предсказания

```python
print('Результаты распознавания: ', decode_predictions(preds, top=3)[0])
```
