```python
from deepface import DeepFace
from deepface.commons import functions
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
```

```python
model1 = "Facenet"
detector1 = "mtcnn"
```

```python
file1 = "/content/drive/MyDrive/number_one.jpg"
file2 = "/content/drive/MyDrive/number_two.jpg"
file3 = "/content/drive/MyDrive/number_three.jpg"
file4 = "/content/drive/MyDrive/number_four.jpg"
file5 = "/content/drive/MyDrive/number_five.jpg"
```

```python
pixels1 = plt.imread(file1)
pixels2 = plt.imread(file2)
pixels3 = plt.imread(file3)
pixels4 = plt.imread(file4)
pixels5 = plt.imread(file5)
```

```python
import mtcnn
```

***Вывод фото***

```python
print("Shape of image/array:",pixels1.shape)
imgplot = plt.imshow(pixels1)
plt.show()
```

Детектирование лица

```python
detected_face1 = DeepFace.detectFace(img_path = file1, detector_backend = detector1, )
plt.imshow(detected_face1)
```

Формируем векторы лица

```python
embedding = DeepFace.represent(img_path = file1, model_name = model1)
print(embedding)
```

______________________________________________________________________________________________________________

```python
print("Shape of image/array:",pixels2.shape)
imgplot = plt.imshow(pixels2)
plt.show()
```

```python
detected_face2 = DeepFace.detectFace(img_path = file2, detector_backend = detector1)
plt.imshow(detected_face2)
```

```python
embedding = DeepFace.represent(img_path = file2, model_name = model1)
print(embedding)
```

```python
result = DeepFace.verify(img1_path = file1, img2_path = file2, model_name=model1, detector_backend=detector1, distance_metric= "euclidean_l2")
print(result)
print("Is verified: ", result["verified"])
```
