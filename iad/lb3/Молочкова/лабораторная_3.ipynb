```python
from deepface import DeepFace
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from keras.preprocessing import image
import numpy as np
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
import matplotlib
from skimage import io
import dlib
import face_recognition
```

```python
img1_path = "/content/drive/MyDrive/Kira.jpg" 
img2_path = "/content/drive/MyDrive/Kira_2.jpg" 
img3_path = "/content/drive/MyDrive/Natali.jpg"
img4_path = "/content/drive/MyDrive/Natali_2.jpg"
```

```python
orig = cv2.imread(img1_path)
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
orig = cv2.imread(img2_path)
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
orig = cv2.imread(img3_path)
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
orig = cv2.imread(img4_path)
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
detected_face = DeepFace.detectFace(img1_path, detector_backend = 'mtcnn')
plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB)) #обнаружение лица
```

```python
detected_face = DeepFace.detectFace(img2_path, detector_backend = 'mtcnn')
plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB))
```

```python
detected_face = DeepFace.detectFace(img3_path, detector_backend = 'mtcnn')
plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB))
```

```python
detected_face = DeepFace.detectFace(img4_path, detector_backend = 'mtcnn')
plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB))
```

```python
result = DeepFace.verify(img1_path, img2_path, distance_metric='euclidean', model_name='Dlib', detector_backend = 'mtcnn')
print( result["verified"])
```
