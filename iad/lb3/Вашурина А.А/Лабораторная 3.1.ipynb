```python
data_dir = '/content/HuH/'
train_dir = 'train'
val_dir   = 'val'
test_dir  = 'test'
test_portion = 0.10
val_portion = 0.10
nb_images = 628
```

```python
def create_dir(dir_name):
  
  if os.path.exists(dir_name):
    shutil.rmtree(dir_name)

  os.makedirs(dir_name)
  os.makedirs(os.path.join(dir_name, "horse"))
  os.makedirs(os.path.join(dir_name, "human"))
```

```python
create_dir(train_dir)
create_dir(test_dir)
create_dir(val_dir)
```

```python
def copy_images(st, ed, src, dst):
  for i in range(st, ed):
    shutil.copy2(os.path.join(src, "horse/" + str(i) + ".png"), os.path.join(dst, "horses"))
    shutil.copy2(os.path.join(src, "human/" + str(i) + ".png"), os.path.join(dst, "humans"))
```

```python
start_val = int(nb_images * (1 - val_portion - test_portion))
start_tst = int(nb_images * (1 - test_portion))
print(start_val)
print(start_tst)
```

```python
copy_images(0, start_val, data_dir, train_dir)
```
