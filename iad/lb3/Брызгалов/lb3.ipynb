```python
def detect_face(img):
    detected_face = DeepFace.detectFace(img, detector_backend = 'dlib')
    plt.imshow(cv2.cvtColor(detected_face, cv2.COLOR_BGR2RGB))
```

```python
def detect_face_orig(img):
    orig = cv2.imread(img)
    plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```

```python
def verify_images(img1, img2):
    obj = DeepFace.verify(img1, img2, distance_metric='euclidean', model_name='OpenFace', detector_backend = 'dlib')
    if (obj['distance'] < 0.78) :
      print(obj['distance'])
      return True
    else :
      return False
```
